```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
p_values = [1, 2, 4, 8, 10, 50]

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots(figsize=(6, 6))
for p in p_values:
    Z = (np.abs(X)**p + np.abs(Y)**p)**(1/p)
    ax.contour(X, Y, Z, levels=[1], colors='b')

ax.set_title(r'Curvas de nivel de la norma $\|X\|_p$')
ax.grid(True)
plt.show()
```

```python
def norm_Q_fixed(x, y):
    vec = np.array([x, y])
    Q = np.array([[3, 1], [1, 10]])
    return np.sqrt(np.dot(vec.T, np.dot(Q, vec)))

Z = np.vectorize(norm_Q_fixed)(X, Y)

plt.figure(figsize=(6, 6))
contour = plt.contour(X, Y, Z, levels=[1], colors='b')
plt.title(r'Curva de Nivel de $\|x\|_Q$')
plt.grid()
plt.show()
```

```python
Q = np.array([[3, 1], [1, 10]])

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

Z = np.sqrt(X**2 * Q[0, 0] + 2 * X * Y * Q[0, 1] + Y**2 * Q[1, 1])

fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')

ax.set_title(r'Superficie de $\|x\|_Q$')

plt.show()
```
